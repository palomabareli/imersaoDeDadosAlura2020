**Pandas**:

Uma das bibliotecas mais utilizadas em Data Sciense. Essa biblioteca é destinada para tratamento de grande volumes de dados. Existem basicamente dois tipos principais de dados nesta biblioteca: Series e DataFrame. 

```
import pandas as pd

#Arquivo com os dados pré-processados (amostra)
file = 'https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true'

#Lendo os dados do CSV e transformando em um Dataframe
data = pd.read_csv(file)

#Exibindo as 5 primeiras linhas, por default
data.head()
```

```
#Analisando a quantidade de linhas e colunas do Dataset
data.shape
```

```
#Selecionando uma das colunas do Dataframe, cuja saída será uma Serie.
data['SG_UF_RESIDENCIA']
```

```
#Selecionando mais de uma coluna do Dataframe, cuja saída será uma Serie.
#Atentar-se que agora os parâmetros são do tipo list!
data[['SG_UF_RESIDENCIA','NO_MUNICIPIO_RESIDENCIA']]
```

```
#Retornando todas as colunas do Dataframe
data.columns.values
```

```
#Selecionando todos os estados da amosta
data['SG_UF_RESIDENCIA'].unique()
```

```
#Selecionando a quantidade de estados da amosta
#Por trás está executando o Numpy,já que é um array
len(data['SG_UF_RESIDENCIA'].unique())
```

```
#Selecionando a quantidade de estados da amosta
data['SG_UF_RESIDENCIA'].value_counts()
```

```
#Agrupa a quantidade de pessoas pela idade da amosta
#Ordenando pelo index
data['NU_IDADE'].value_counts().sort_index()
```

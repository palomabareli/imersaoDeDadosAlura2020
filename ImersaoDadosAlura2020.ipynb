**Pandas**:

Uma das bibliotecas mais utilizadas em Data Sciense. Essa biblioteca é destinada para tratamento de grande volumes de dados. Existem basicamente dois tipos principais de dados nesta biblioteca: Series e DataFrame. 

```
import pandas as pd

#Arquivo com os dados pré-processados (amostra)
file = 'https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true'

#Lendo os dados do CSV e transformando em um Dataframe
data = pd.read_csv(file)

#Exibindo as 5 primeiras linhas, por default
data.head()
```

```
#Analisando a quantidade de linhas e colunas do Dataset
data.shape
```

```
#Selecionando uma das colunas do Dataframe, cuja saída será uma Serie.
data['SG_UF_RESIDENCIA']
```

```
#Selecionando mais de uma coluna do Dataframe, cuja saída será uma Serie.
#Atentar-se que agora os parâmetros são do tipo list!
data[['SG_UF_RESIDENCIA','NO_MUNICIPIO_RESIDENCIA']]
```

```
#Retornando todas as colunas do Dataframe
data.columns.values
```

```
#Selecionando todos os estados da amosta
data['SG_UF_RESIDENCIA'].unique()
```

```
#Selecionando a quantidade de estados da amosta
#Por trás está executando o Numpy,já que é um array
len(data['SG_UF_RESIDENCIA'].unique())
```

```
#Selecionando a quantidade de estados da amosta
data['SG_UF_RESIDENCIA'].value_counts()
```

```
#Agrupa a quantidade de pessoas pela idade da amosta
#Ordenando pelo index
data['NU_IDADE'].value_counts().sort_index()
```

Existem outras formas, além de tabelas para reinterpretação de dados visuais.
Para isso são usados gráficos.
Os gráficos aqui estão usando por baixo dos panos a biblioteca **Matplotlib**.

```
#utilizando o gráfico de Histograma para visulizar os dados de idade
#cuidado com o padrão dos ranges para não gerar um gráfico confuso!
data['NU_IDADE'].hist()
```

```
#utilizando o gráfico de Histograma para visulizar os dados de idade
#só que agora customizando o intervalo para 20
#e o tamanho da imagem para 10 por 8
data['NU_IDADE'].hist(bins=20, figsize = (10,8))
```

```
#Treineiro: são as pessoas que participam do processo
#mas sua nota não é contabilizada para notas
#das faculdades. Neste caso,estamos selecionando 
#somente os treineiros

#Retonar um Dataframe
data.query('IN_TREINEIRO == 1')

#Para conferencia dos dados
#Retorna uma Serie
#data.query('IN_TREINEIRO == 1')['IN_TREINEIRO']
```

```
#Treineiros por idade
#Retorna uma Serie
data.query('IN_TREINEIRO == 1')['NU_IDADE'].value_counts().sort_index
```

```
#Histograma por nota de Redação
data['NU_NOTA_REDACAO'].hist(bins = 20, figsize = (8,6))
```

```
#Estatística por média de nota de Redação
data['NU_NOTA_REDACAO'].mean()
```

```
#Estatística por desvio padrão de nota de Redação
data['NU_NOTA_REDACAO'].std()
```

```
#Selecionando somente as variáveis de interesse
#para ser utilizada em filtro
test = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']

#Descrição Estatística
data[test].describe()

#Função para os quartis -> os 10% mais alto da amostra
#data['NU_NOTA_CN'].quantile(0.9)
```

```
#Descrição Estatística via gráfico Boxplot
data['NU_NOTA_CN'].plot.box(grid = True, figsize=(8,6))
```

```
#Gráfico Boxplot de todas as notas
data[test].boxplot(grid = True, figsize=(8,6))
#Explicando:
#NU_NOTA_LC = concentrou mais indivíduos com a mesma notas próximas

#Medidas de Resumo
#Média Central: tendência a medir o centro
#Medida de dispersão: como estes dados estão distribuidos na amostra
```
